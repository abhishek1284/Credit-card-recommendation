```python

# --- 1. Distribution Analysis ---
plt.figure(figsize=(10, 5))
sns.histplot(data['person_emp_length'].dropna(), bins=20, kde=True)
plt.title('Distribution of Employment Length (Years)')
plt.xlabel('Employment Length (Years)')
plt.ylabel('Frequency')
plt.show()

# --- 2. Binning the Employment Length ---
bins = [0, 2, 5, 10, 20, data['person_emp_length'].max()]
labels = ['0-2 yrs', '3-5 yrs', '6-10 yrs', '11-20 yrs', '21+ yrs']
data['emp_length_bin'] = pd.cut(data['person_emp_length'], bins=bins, labels=labels, right=False)

# --- 3. Impact on Default Rate ---
# Assume 'loan_status' is 1 for default, 0 for paid
default_rate = data.groupby('emp_length_bin')['loan_status'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='emp_length_bin', y='loan_status', data=default_rate)
plt.title('Default Rate by Employment Length Bin')
plt.ylabel('Default Rate')
plt.xlabel('Employment Length Bin')
plt.ylim(0, 1)
plt.show()

# --- 4. Correlation with Loan Status ---
correlation = data[['person_emp_length', 'loan_status']].corr()
print("Correlation between employment length and loan status:")
print(correlation)

# Optional: Show default rate and count for each bin
summary = data.groupby('emp_length_bin')['loan_status'].agg(['mean', 'count']).reset_index()
print(summary)
```

```python

# --- 1. Default Rate by Loan Intent ---
default_rate = data.groupby('loan_intent')['loan_status'].mean().reset_index()

plt.figure(figsize=(10, 5))
sns.barplot(x='loan_intent', y='loan_status', data=default_rate, palette='Set2')
plt.title('Default Rate by Loan Intent')
plt.xlabel('Loan Intent')
plt.ylabel('Default Rate')
plt.ylim(0, 1)
plt.xticks(rotation=45)
plt.tight_layout()
plt.savefig("loan.png")
plt.show()

# --- 2. Frequency and Average Loan Amount ---
loan_stats = data.groupby('loan_intent')['loan_amnt'].agg(['count', 'mean']).reset_index()

fig, ax1 = plt.subplots(figsize=(12, 6))

sns.barplot(x='loan_intent', y='count', data=loan_stats, ax=ax1, color='skyblue')
ax1.set_ylabel('Frequency', color='blue')
ax1.set_xlabel('Loan Intent')
ax1.set_title('Frequency and Average Loan Amount by Loan Intent')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_xticklabels(loan_stats['loan_intent'], rotation=45)

# Create a second y-axis for the average loan amount
ax2 = ax1.twinx()
sns.lineplot(x='loan_intent', y='mean', data=loan_stats, ax=ax2, color='red', marker='o')
ax2.set_ylabel('Average Loan Amount', color='red')
ax2.tick_params(axis='y', labelcolor='red')

plt.tight_layout()
plt.savefig("freq.png")

plt.show()

# --- 3. One-Hot Encoding for Machine Learning ---
# Create dummy variables for 'loan_intent'
loan_intent_dummies = pd.get_dummies(data['loan_intent'], prefix='intent', drop_first=False)

# Concatenate with the original dataset
data_encoded = pd.concat([data, loan_intent_dummies], axis=1)

# Display result
print("One-hot encoded columns added:")
print(data_encoded.columns)
```
